# stratify patients by primary tumor

```python
tumortypeDict = {}
primarytumorEnrollementDict = {}

metEnrollmentDict = {}
metEnrollmentPresence = {}

for ptID in ptDict.keys():
    # tumortype = ptDict[pt].primarytumor
    pt = ptDict[ptID]
    if pt.ptDemographics is not False:
        tumortypeDict[ptID] = pt.primarytumor #numeric code for type of primary tumor
        primarytumorEnrollementDict[ptID]= pt.primarytumorEnrollment #binary if primary tumor is present at enrollment
        metEnrollmentDict[ptID] = pt.metEnrollmentList #numeric list of locations of mets at enrollment 
        metEnrollmentPresence[ptID] = pt.metsAtEnrollment #length of met sites
```

```python
#cluster ptIDs (keys) from metEnrollmentDict by similar values in list
metlocations = [1,2,3,4,5,6]
metSort = {} #keys are met locations, values are patient IDs with that met location at enrollment. may have other mets
metSortSingle = {} #like above, but only has the one met location

for location in metlocations: 
    metLocationTemp = [] #list of patient IDs that have met at enrollment at a given location
    metSingleTemp = [] #list of patient IDs where this location is the only met location at enrollment
    for ptID in metEnrollmentDict.keys():
        if location in metEnrollmentDict[ptID]:
           metLocationTemp.append(ptID)
           if len(metEnrollmentDict[ptID]) == 1: 
              metSingleTemp.append(ptID)
    metSort[location] = metLocationTemp
    metSortSingle[location]= metSingleTemp
```
